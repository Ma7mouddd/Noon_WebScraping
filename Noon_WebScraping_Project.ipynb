```python
import re
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd
import numpy as np
from pyvirtualdisplay import Display
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.keys import Keys

# Start virtual display
display = Display(visible=0, size=(800, 600))
display.start()

# Set up the WebDriver
driver = webdriver.Firefox()
page_url = "https://www.noon.com/egypt-en/"
driver.get(page_url)

# Write in search bar "Mobiles"
search_bar = driver.find_element(By.XPATH, "//input[@id='searchBar']")
search_bar.send_keys("Mobiles")
search_bar.send_keys(Keys.ENTER)
time.sleep(3)

items = []
while True:
    # Extract all devices
    titles = driver.find_elements(By.XPATH, "//div[@class='sc-26c8c6bb-24 cCbHzm']")
    reviews = driver.find_elements(By.XPATH, "//div[@class='sc-2709a77c-2 hUinXQ']")
    prices = driver.find_elements(By.CLASS_NAME, "amount")
    # Loop to get the title, review, and price for each device
    for title, review, price in zip(titles, reviews, prices):
        try:
            print("Title:", title.text)
            print("Review:", review.text)
            print("Price:", price.text)
            print("-" * 40)
            # Append the data to the items list
            items.append([title.text, review.text, price.text])
        except Exception as e:
            print("Error:", e)

    # Check for the presence of the Next button
    try:
        next_button = driver.find_element(By.XPATH, "//a[@aria-label='Next page']")
        next_button.click()
        time.sleep(3)
    except Exception as e:
        print("No more pages available or an error occurred.")
        break

# Convert the data to a pandas DataFrame
data = pd.DataFrame(items, columns=["MobileOrAccessories", "Review", "Price"])

# Optionally save the data to a CSV file
data.to_csv("noon_mobiles.csv", index=False)

# Close the WebDriver and stop the virtual display
driver.quit()
display.stop()
```
